# Round 1

## Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
from sklearn import linear_model
import statsmodels.api as sm
%matplotlib inline
```

## Read the files 

```python
df = pd.read_csv('train.csv')
```

```python
df.head(10)
```

## Dropping all the features except the 5 needed

```python
df_train = df[['Year_Factor', 'building_class', 'avg_temp', 'site_eui', 'id']]
df_train.head()
```

## Check for missing values 

```python
df_train.info()     # No null values
```

```python
df_train.describe()
```

```python
for i in (['Year_Factor', 'building_class']):
    print(df_train[i].unique())
```

## Replacing categorical values to numbers

```python
df_train['building_class'].replace(['Commercial', 'Residential'], [0,1], inplace=True)
df_train.head()
```

```python
df_train.tail(10)
```

## Building the model 

```python
X = df_train[['Year_Factor', 'building_class', 'avg_temp']]
y = df_train['site_eui']

model = linear_model.LinearRegression()
training_model = model.fit(X,y)
```

# Preparing the test data

```python
df2 = pd.read_csv('test.csv')
df_id = df2['id']
df_id = pd.DataFrame(df_id, columns = ['id'])
df_id.head()
```

```python
df_test = df2[['Year_Factor', 'building_class', 'avg_temp', 'id']]
df_test.head()
```

```python
df_test['building_class'].replace(['Commercial', 'Residential'], [0,1], inplace=True)
```

```python
df_test.head()
```

## Predicting the test results 

```python
df_test.drop('id', axis=1, inplace=True)
y_test = model.predict(df_test)
```

```python
y_pred = pd.DataFrame(y_test, columns = ['site_eui'])
y_pred
```

```python
result = df_id.combine_first(y_pred)
```

```python
result.head()
```

```python
result.to_csv('submissions.csv', index=False)
```
